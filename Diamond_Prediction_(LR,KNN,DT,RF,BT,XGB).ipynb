```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from scipy import stats
import math


path = "/content/drive/MyDrive/INNFO367/Week 8/diamonds_train.csv"
df = pd.read_csv(path)

df.head()
```

```python
df = df.sample(n=30000, random_state= 2)

df.head()
```

```python
df.info()
```

```python
df = pd.get_dummies(df, drop_first=True)

df.head()
```

```python
X = df.drop('price', axis=1)     # Features (independent variables)
y = df['price']

X.head()
```

```python
train_X, valid_X, train_y, valid_y = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
re_prices = LinearRegression()

re_prices.fit(train_X, train_y)
```

```python
train_X.head()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, max_error, r2_score, explained_variance_score
```

```python
pred_y = re_prices.predict(valid_X)
```

```python
print('Regression model performance evaluation')
print('R2: ', r2_score(valid_y, pred_y))
print('MAE: ', mean_absolute_error(valid_y, pred_y))
print('RMSE: ', math.sqrt(mean_squared_error(valid_y, pred_y)))
print('Max error: ', max_error(valid_y, pred_y))
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_ = scaler.fit_transform(X)

X_rescaled = pd.DataFrame(X_, columns=X.columns)
```

```python
X_rescaled.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_rescaled, y, test_size=0.3, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

knn = KNeighborsRegressor(n_neighbors=5, metric='euclidean')
knn.fit(X_train, y_train)

# Make predictions for regression
y_pred = knn.predict(X_test)

# Calculate regression metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
max_err = max_error(y_test, y_pred)


# Print the regression metrics
print("R-squared (R2) Score:", r2)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Maximum Error:", max_err)
```

```python
from sklearn.model_selection import cross_val_score

max_K = 50
cv_scores = []

for K in range(1, max_K):
    knn = KNeighborsRegressor(n_neighbors=K)
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring="r2")
    cv_scores.append(scores.mean())

print('Optimal k: ', cv_scores.index(max(cv_scores))+1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

knn = KNeighborsRegressor(n_neighbors=3, metric='euclidean')
knn.fit(X_train, y_train)

# Make predictions for regression
y_pred = knn.predict(X_test)

# Calculate regression metrics
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
max_err = max_error(y_test, y_pred)


# Print the regression metrics
print("R-squared (R2) Score:", r2)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Maximum Error:", max_err)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

dt_model = DecisionTreeRegressor(min_samples_leaf=5, max_depth=3)
dt_model.fit(X_train, y_train)

# Make predictions for regression
y_pred = dt_model.predict(X_test)

# Calculate regression metrics
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
max_err = max_error(y_test, y_pred)

# Print the regression metrics
print("R-squared (R2) Score:", r2)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Maximum Error:", max_err)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

rf_model = RandomForestRegressor(max_depth=5, random_state=0)
rf_model.fit(X_train, y_train)

# Make predictions for regression
y_pred_rf = rf_model.predict(X_test)

# Calculate regression metrics
mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
r2_rf = r2_score(y_test, y_pred_rf)
max_err_rf = max_error(y_test, y_pred_rf)

# Print the regression metrics
print("R-squared (R2) Score for Random Forest:", r2_rf)
print("Mean Absolute Error (MAE) for Random Forest:", mae_rf)
print("Root Mean Squared Error (RMSE) for Random Forest:", rmse_rf)
print("Maximum Error for Random Forest:", max_err_rf)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

bt_model = AdaBoostRegressor(n_estimators=100)
bt_model.fit(X_train, y_train)

# Make predictions for regression
y_pred_bt = bt_model.predict(X_test)

# Calculate regression metrics
mae_bt = mean_absolute_error(y_test, y_pred_bt)
rmse_bt = mean_squared_error(y_test, y_pred_bt, squared=False)
r2_bt = r2_score(y_test, y_pred_bt)
max_err_bt = max_error(y_test, y_pred_bt)

# Print the regression metrics
print("R-squared (R2) Score for Boosted Tree:", r2_bt)
print("Mean Absolute Error (MAE) for Boosted Tree:", mae_bt)
print("Root Mean Squared Error (RMSE) for Boosted Tree:", rmse_bt)
print("Maximum Error for Boosted Tree:", max_err_bt)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Create and train an XGBoost regression model
xgb_model = xgb.XGBRegressor()
xgb_model.fit(X_train, y_train)

# Make predictions for regression
y_pred_xgb = xgb_model.predict(X_test)

# Calculate regression metrics
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
rmse_xgb = mean_squared_error(y_test, y_pred_xgb, squared=False)
r2_xgb = r2_score(y_test, y_pred_xgb)
max_err_xgb = max_error(y_test, y_pred_xgb)

# Print the regression metrics
print("R-squared (R2) Score for XGBoost:", r2_xgb)
print("Mean Absolute Error (MAE) for XGBoost:", mae_xgb)
print("Root Mean Squared Error (RMSE) for XGBoost:", rmse_xgb)
print("Maximum Error for XGBoost:", max_err_xgb)
```
